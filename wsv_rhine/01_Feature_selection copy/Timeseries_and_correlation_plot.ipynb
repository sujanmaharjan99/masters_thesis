```python
import os
import pandas as pd

from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
directory_list = [os.path.join('..', 'Inputs_outputs_excels', 'wsv_rhine_export - pseudo_huber_29_4_2024.xlsx')]

df_pseudo = pd.read_excel(directory_list[0])

df_pseudo.index = pd.to_datetime(df_pseudo['Date'],format='%d.%m.%Y')
```

```python
start_date='15/10/2002'
end_date='30/12/2002'

start_date_obj = datetime.strptime(start_date, '%d/%m/%Y')
end_date_obj = datetime.strptime(end_date, '%d/%m/%Y')

fig, (ax1) = plt.subplots(figsize=(9, 5))
ax2 = ax1.twinx()

ax1.plot(df_pseudo.index, df_pseudo['BONN/flow.obs'], label='Bonn - Obs', color='red')

ax2.plot(df_pseudo.index, df_pseudo['BONN/flow.bc'], label='Bonn - MBC', color='blue')

ax1.set_title('Discharge correlation with Mass balance correction ', fontsize=14, fontweight='bold')
ax1.set_ylabel('Discharge [m³/s]', fontsize=12)
ax1.set_xlim(start_date_obj, end_date_obj)
ax1.set_ylim(1000, 6500)
ax1.grid(True)
ax1.legend(loc='upper left')

ax2.set_ylabel('MBC [m³/s]', fontsize=12)
ax2.set_xlim(start_date_obj, end_date_obj)
ax2.set_ylim(-400,100)
#ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.savefig('correlation_plt1.png')
plt.show()
```
